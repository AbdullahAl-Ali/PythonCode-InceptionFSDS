# Nested Conditional Statement


```python
# Login function demo with conditional statements - pythonwithabdullah.com
"""
Abdutal  Credentionals:
username: abdullahalali.c@gmail.com
password: abdullah@12345
"""
email = input("Enter your email: ")
password = input("Enter your password: ")


actual_email = "abdullahalali.c@gmail.com"
actual_password = "abdullah@12345"

if email == actual_email and password == actual_password:
    print("Log in Sucsessful!")

elif email == actual_email and password != actual_password:
    print("Incorrect Password, Please try again!")
    password = input("Enter your password again: ")

    if password == actual_password:
        print("Log in Sucsessful!")

    else:
        print("Incorrect Password, Account locked!")
        print("Please contact admistrator to reset your password.")
        

elif email != actual_email and password != actual_password:
    print("Email doesn't match with the password!")

else:
    print("Invalid Credential.")
```
