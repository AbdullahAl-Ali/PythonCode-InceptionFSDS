# Operations on Strings

- Arithmetic Operations
- Relational Operations
- Logical Operations
- Loops on Strings
- Membership Operations

```python
# Arithmetic Operations
print("Hi" + " " + " Everyone") #concatination
```

```python
print("Python" * 3)  # repetition
print("Hello" - "l")  # This will raise an error because subtraction is not supported for strings
```

```python
print("*" * 5)  # prints *****
```

```python
# Relational Operations

print("Hello" == "Hello") # equality
print ( "Hello" != "World") # not equal
print ("Hello" > "Hi") # greater than
print ("Hello" < "World") # less than
print ("Hello" >= "Hello") # greater than or equal to
```

```python
# Logical Operations

"hello" and 'world' # returns 'world', because both are true values, so it returns the last value
```

```python
"Hello" or "world "  # returns 'Hello', because the first value is true, so it returns the first value
```

```python
not "Hello"  # returns False, because "Hello" is a non-empty string which is considered True
```

```python
for i in "Abdullah":
    print(i)
```

```python
# Membership Operations

print('a' in 'Abdullah')  # True, because 'a' is present in 'Abdullah'
print('z' not in 'Abdullah')  # True, because 'z' is not present in 'Abdullah'
print('X' in 'Abdullah')  # False, because 'X' is not present in 'Abdullah'
print('b' not in 'Abdullah')  # False, because 'b' is present in 'Abdullah'
```
