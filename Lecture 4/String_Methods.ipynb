# Common Functions/Methods
- len
- max
- min
- sorted


```python
s = "Welcome to Inception FSDS"
```

```python
max(s)  # returns 't', because 't' has the highest ASCII value among all characters in the string
```

```python
min(s)  # returns ' ' (space), because space has the lowest ASCII value among all characters in the string
```

```python
sorted(s) # returns a list of characters in ascending order based on their ASCII values
```

```python
sorted(s, reverse=True) # returns a list of characters in descending order based on their ASCII values
```

```python
s = "Welcome to Inception FSDS"
```

```python
s.capitalize()  # 'Welcome to inception fsds' - Capitalizes the first character of the string
```

```python
s.upper()  # 'WELCOME TO INCEPTION FSDS' - Converts all characters to uppercase
```

```python
s.lower()  # 'welcome to inception fsds' - Converts all characters to lowercase
```

```python
s = "Welcome to Inception FSDS"
```

```python
s.swapcase()  # 'wELCOME TO iNCEPTION fsds' - Swaps the case of each character in the string
```

- Count
- Find
- Index


```python
s = "My name is Abdullah. I am learning Python programming. Python is great for data science."
```

```python
s.count("Python")  # 2 - Counts occurrences of the substring "Python"
```

```python
s.find("A")  # 11 - Finds the first occurrence of substring "A", returns -1 if not found
```

```python
s.find("x")  # -1 - 'x' is not found in the string
```

```python
s.index("A")  # 11 - Finds the first occurrence of substring "A", raises ValueError if not found
```

```python
s.("x")  # Raises ValueError - 'x' is not found in the string

```

```python
s = "My name is Abdullah"
```

```python
s.startswith("My")  # True - Checks if the string starts with "My"
```

```python
s.endswith("ah")  # True - Checks if the string ends with "ah"
```

### Format

```python
name = "Abdullah"
gender = "Male"
age = 25

s = "My name is {} and I am a {}. I am {} years old."
```

```python
s.format(name, gender, age) # 'My name is Abdullah and I am a Male. I am 25 years old.' - puts the variables in the specified order
```

```python
name = "Abdullah"
gender = "Male"
age = 25

s = "My name is {2} and I am a {1}. I am {0} years old."     # Using positional arguments in format
```

```python
s.format(age, gender, name) # puts the variables in the specified order
```

# F-string

```python
s = f"My name is {name} and I am a {gender}. I am {age} years old."  # puts the variables directly in the string
print(s)
```

- isalnum

```python
s = "Abdullah@123"
s.isalnum()  # True - Checks if all characters are alphanumeric
```

```python
s = "123456"
s.isalnum()  # True - Checks if all characters are alphanumeric
```

```python
s = "Abdullah"
s.isalpha()  # True - Checks if all characters are alphabetic 
```

```python
s = "123456"
s.isdigit()  # True - Checks if all characters are digits
```

```python
s = "variable_name_1"
s.isidentifier()  # True - Checks if the string is a valid identifier
```

• Split
• Join

```python
s = "My, name, is, Abdullah"
```

```python
s.split(",") # ['My', ' name', ' is', ' Abdullah'] - Splits the string at each comma and returns a list
```

```python
s = "My@name@is@Abdullah"
```

```python
s.split("@")
```

```python
# Join

words = ['My', 'name', 'is', 'Abdullah']

```

```python
s = " "  # Empty string as separator
s = s.join(words)
print(s)
```

• Replace

```python
s = "My name is Abdullah"
```

```python
s.replace("Abdullah", "Ali")  # 'My name is John' - Replaces occurrences of "Abdullah" with "Ali"
```

```python
s = "  My name is Abdullah     "
```

```python
s = s.strip() # 'My name is Abdullah' - Removes leading and trailing whitespace
```
