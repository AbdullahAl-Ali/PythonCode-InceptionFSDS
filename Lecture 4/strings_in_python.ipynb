# Accessing Strings


```python
s = "welcome to python"
```

```python
print(s[0]) # first character
print(s[4])  # fifth character
print(s[11]) # twelfth character
print(s[-1]) # last character
print(s[-3]) # third last character
```

```python
len(s)
```

```python
# Negetive Indexing

s = "welcome to python"
print(s[-1]) # last character
print(s[-2]) # second last character
print(s[-11]) # eleventh last character
```

```python
# Positive slicing
s = "welcome to python"
print(s[11:17])
print(s[11:])
print(s[:17])
print(s[11:10007])
```

```python
s[11:17]  
```

```python
s = "welcome to python"
```

```python
print(s[0:7:3])  
```

```python
print(s[7::-1]) # reverse from index 7 to 0
print(s[::-1]) # full reverse
print(s[::-2]) # reverse with 1 character skip
```
